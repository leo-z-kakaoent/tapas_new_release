```python
import pickle
import d3rlpy
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from tqdm import tqdm
from sklearn.metrics import accuracy_score, ndcg_score
from torch.utils.data import Dataset, DataLoader
from scipy.sparse import coo_matrix, csr_matrix
pd.options.mode.chained_assignment = None
```

```python
def hard_profit(y_score, y_true):
    return np.sum(y_true[np.where(y_score>0)])
    
def naive_profit(y_score, y_true):
    return np.sum(y_true[np.where((y_score>0)|(y_true>0))])

def mean_profit(y_score, y_true):
    hard = hard_profit(y_score, y_true)
    naive = naive_profit(y_score, y_true)
    return np.mean([hard, naive])

def accuracy(y_score, y_true):
    bin_pred = np.where(y_score>0, 1, 0)
    bin_true = np.where(y_true>0, 1, 0)
    return accuracy_score(y_pred=bin_pred, y_true=bin_true)

```

```python
y_true = resultdf.true_profit.values
rlist = []

for model in ['default','mf','lightfm','bcq','cql']:
    y_score = resultdf['%s_score'%model].values
    acc = accuracy(y_score=y_score, y_true=y_true)
    hardp = hard_profit(y_score=y_score, y_true=y_true)
    naivep = naive_profit(y_score=y_score, y_true=y_true)
    meanp = mean_profit(y_score=y_score, y_true=y_true)
    # top1p = topp(model, 100)
    # top10p = topp(model, 10)
    # top50p = topp(model, 2)
    rlist.append({
        'model': model,
        'accuracy': acc,
        'hard profit': hardp * 1000*1.2/10000,
        'naive profit': naivep* 1000*1.2/10000,
        'mean profit': meanp* 1000*1.2/10000,
        # 'top 1% accuracy': top1p,
        # 'top 10% accuracy': top10p,
        # 'top 50% accuracy': top50p,
    })
pd.DataFrame().from_dict(rlist)
```

```python
def topp(model, p):
    truep = resultdf.groupby('series_id').apply(lambda x: x.sort_values('true_profit')['user_id'].head(int(len(x)/p)).values)
    modelp = resultdf.groupby('series_id').apply(lambda x: x.sort_values('%s_score'%model)['user_id'].head(int(len(x)/p)).values)
    t = 0
    s = 0
    for k,v in truep.to_dict().items():
        t += len(v)
        s += np.sum(np.isin(modelp[k],v))
    return s/t
```
